## Finally give the feedback for all movies with ‘Good’ if the movie ‘rating’ is >= 8 and ‘Average’ if the movie rating doesn’t follow the condition given.

```python
def RatingReview(x):
    if x >= 9.0:
        return "Great"
    elif x >= 8.0:
        return "Good"
    elif x >= 7.0:
        return "Fairly Good"
    elif x >= 6.0:
        return "Average"
    elif x >= 5.0:
        return "Just ok"
    elif x >= 4.0:
        return "Pretty Bad"
    elif x >= 3.0:
        return "Bad"
    elif x >= 2.0:
        return "Awful"
    else:
        return "Terrible"
```

```python
df["rating_review"] = df['rating'].apply(RatingReview)
```
